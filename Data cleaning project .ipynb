```python
import re

def clean_name(x):
    return re.sub(r"[^\w\s]", "", str(x))

df["clean_last_name"] = df["Last_Name"].apply(clean_name)
```

```python
df["clean_last_name"]=df["Last_Name"].apply(clean_name)
```

```python
df
```

```python
def clean_name1(x):
    return re.sub(r"[^\w\s]", "", str(x)).replace("_", "")
```

```python
df["clean_last_name"]=df["clean_last_name"].apply(clean_name1)
df
```

```python
def clean_number(x):
    return re.sub(r"[^0-9]","",str(x))
```

```python
df["clean_Phone_Number"]=df["Phone_Number"].apply(clean_number)
df
```
